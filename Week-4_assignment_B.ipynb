#### - Generation of an iteration of Q(d) for d = [1, dmax]

```python
import random, math
import numpy as np
    
def V_sph(dim):
    return math.pi ** (dim / 2.0) / math.gamma(dim / 2.0 + 1.0)
```

```python
V_sph(1)
```

```python
d_max = 200
Q_d = []
for d in range(1, d_max-1):
    Q = V_sph(d+1)/V_sph(d)
    Q_d.append(Q)
print(Q_d)
```

#### - Comparision of V_sp(4) with pi^2/2

```python
print("V_sph(4) numerically = ", V_sph(4))

print("pi^2/2 = ", (math.pi **2)/2)
```

The numbers are close match

#### - Calculation of V_sph(200)

```python
d_max = 200
Q0 = 2.0
for d in range(1, d_max):
    Q1 = V_sph(d+1)/V_sph(d)
    V_200 = Q0 * Q1
print("V_sph(200) = ", V_200)
```

```python
d_max = 200
V_d1 = []
d1 = []
for d in range(1, d_max):
    V_d = V_sph(d)
    V_d1.append(V_d)
    d1.append(d)
    
```
